```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# df = pd.read_csv('foreign_names.csv',sep=";")
# df.columns = ['id', 'name', 'meaning', 'gender', 'origin', 'PeoplesCount', 'WhenPeoplesCount', 'Source']

#Lcount = 25
# DEC = 97

Lcount = 31
DEC = 1072
df = pd.read_csv('russian_names.csv',sep=";")
df.columns = ['ID', 'Name', 'Sex', 'PeoplesCount', 'WhenPeoplesCount', 'Source']
df = df.rename(columns={"ID": "id", 
                    "Name": "name",
                    "Sex": "gender",
                  })
df.head()
```

```python
# df = df.drop(['id', 'meaning', 'origin', 'PeoplesCount', 'WhenPeoplesCount', 'Source'], 1)
df = df.drop(['id', 'PeoplesCount', 'WhenPeoplesCount', 'Source'], 1)
df = df.dropna(axis = 0, how ='any') 
df = df.iloc[np.random.permutation(len(df))]
df.head()
```

```python
df.drop(df.loc[df['name'].str.len() < 2].index, inplace = True)
df['name'] = df['name'].str.lower()
df.head()
```

```python
mapping = {'Male': 1, 'Female': 0}
mapping = {'М': 1, 'Ж': 0}
df = df.replace({'gender': mapping})
df.head()
```

```python
df['nameStart']   = df['name'].str[:2]
df['nameEnd']   = df['name'].str[-2:]

df['nameStartValue'] = df.groupby('nameStart').gender.transform(np.mean)*2-1
df['nameEndValue'] = df.groupby('nameEnd').gender.transform(np.mean)*2-1
#df['gender'] = df['gender']*2-1
df.head()
```

```python
Xdata = df.drop(['name', 'gender', 'nameStart', 'nameEnd'], 1)
Ydata = df['gender']
```

```python
logistic = LogisticRegression()
X = Xdata
Y = Ydata
logistic.fit(X, Y)
logistic.score(X, Y)
print('Coefficient: \n', logistic.coef_)
print('Intercept: \n', logistic.intercept_)
print('R² Value: \n', logistic.score(X, Y))
```
